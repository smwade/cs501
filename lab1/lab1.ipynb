```python
import seaborn
import numpy as np
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
def randomLine():
    width = np.random.randint(15)
    y_inter = np.random.randint(20)
    slope = np.random.uniform(-10,10)
    x = np.linspace(-10,10,500)
    y = lambda x: slope * x + y_inter
    
    plt.plot(x, y(x))
```

```python
def randomPoint():
    point = np.random.randint(-10,10,size=2)
    plt.scatter(point[1],point[0])
```

```python
for x in xrange(20):
    randomLine()
    randomPoint()
plt.axis('off')
plt.show()
```
