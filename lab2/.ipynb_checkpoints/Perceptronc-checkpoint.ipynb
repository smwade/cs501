```python
import numpy as np
import pandas
from matplotlib import pyplot as plt
```

```python
def unpickle(file):
    import cPickle
    fo = open(file, 'rb')
    dict = cPickle.load(fo)
    fo.close()
    return dict
```

```python
data = pandas.read_csv( 'Fisher.csv' )
m = data.as_matrix()
labels = m[:,0]
labels[ labels==2 ] = 1  # squash class 2 into class 1
labels = np.atleast_2d( labels ).T
features = m[:,1:5]


data2 = unpickle( 'cifar-10-batches-py/data_batch_1' )
 
features2 = data2['data']
labels2 = data2['labels']
labels2 = np.atleast_2d( labels2 ).T

# squash classes 0-4 into class 0, and squash classes 5-9 into class 1
labels2[ labels2 < 5 ] = 0
labels2[ labels2 >= 5 ] = 1
```

```python
class Perceptron:
    def __init__(self, D, L, c):
        self.data = D
        self.data = np.hstack((self.data, np.ones((self.data.shape[0],1))))
        self.labels = L.flatten()
        self.c = c
        self.w = np.random.normal(0,.001,self.data.shape[1])
        self.cat = np.zeros(len(L))
        self.acc_hist = []
        self.norm_hist = []

    def f(self, x):
        return (np.dot(self.w,x) > 0)

    def train(self, iters=100):
        for i in xrange(iters):
            for j in xrange(len(self.data)):
                y = self.f(self.data[j])
                self.cat[j] = y
                self.w = self.w + self.c*(self.labels[j] - y)*self.data[j]
            
            accuracy = (float(len(self.cat)) - np.sum(np.abs(self.cat - self.labels))) / float(len(self.cat))
            self.acc_hist.append((float(len(self.cat)) - np.sum(np.abs(self.cat - self.labels)))/float(len(self.cat)))
            
            norm = np.sqrt(np.dot(self.w, self.w))
            self.norm_hist.append(norm)
            
    def test(self, sample):
        return self.f(sample)
```

```python
p1 = Perceptron(features, labels, 1.0) 
p1.train()
print p1.w

p2 = Perceptron(features, labels, 0.1) 
p2.train()
print p2.w

p3 = Perceptron(features, labels, 0.01) 
p3.train()
print p3.w

t = np.linspace(1,100,100)

plt.figure(figsize=(9,9))
plt.plot(t, p1.acc_hist, label='c=1')
plt.plot(t, p2.acc_hist, label='c=0.1')
plt.plot(t, p3.acc_hist, label='c=0.01')
plt.xlabel('Iterations')
plt.ylabel('Classification Accuracy')
plt.legend(loc=0)
plt.show()

plt.figure(figsize=(9,9))
plt.plot(t, np.log(p1.norm_hist), label='c=1')
plt.plot(t, np.log(p2.norm_hist), label='c=0.1')
plt.plot(t, np.log(p3.norm_hist), label='c=0.01')
plt.xlabel('Iterations')
plt.ylabel('Natural Logarithm of L2 Norm of Weight Vector')
plt.legend(loc=0)
plt.show()
```

```python
p4 = Perceptron(features2, labels2, 0.001) 
p4.train()
print p4.w

p5 = Perceptron(features2, labels2, 0.00001) 
p5.train()
print p5.w

plt.figure(figsize=(9,9))
plt.plot(t, p4.acc_hist, label='c=0.001')
plt.plot(t, p5.acc_hist, label='c=0.00001')
plt.xlabel('Iterations')
plt.ylabel('Classification Accuracy')
plt.legend(loc=0)
plt.show()

plt.figure(figsize=(9,9))
plt.plot(t, np.log(p4.norm_hist), label='c=0.001')
plt.plot(t, np.log(p5.norm_hist), label='c=0.00001')
plt.xlabel('Iterations')
plt.ylabel('Natural Logarithm of L2 Norm of Weight Vector')
plt.legend(loc=0)
plt.show()
```
